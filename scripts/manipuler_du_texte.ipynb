## Niveau 1 : présence d'un mot dans un texte

Souvent, on veut vérifier la présence (ou l'absence) d'un mot dans un texte. Sur word ou excel, on utilise la fonction "rechercher/" (`ctrl+f`)

Sur python, plusieurs methodes sont possibles. Nous allons en voir 2 :

- La première utilise les fonctions natives de Python
- la seconde suppose de faire appel à une librairie


**On commence par définir un texte**

### Méthode 1 : recherche à l'intérieur d'une liste de mot

La première méthode consiste à découper le texte en liste de mot avec la fonction `.split()`, puis à regarder si chacun des mots correspon à celui recherché

````python
texte = """Alice, assise auprès de sa sœur sur le gazon, commençait à s'ennuyer de rester là à ne rien faire ; 
une ou deux fois elle avait jeté les yeux sur le livre que lisait sa sœur ; mais quoi ! 
pas d'images, pas de dialogues ! « La belle avance, » pensait Alice, « qu’un livre sans images, sans causeries ! »"""
````

````python
compteur = 0 #on ajoutera +1 au compteur à chaque fois que l'on trouve le mot
mot_recherche = "livre" #le mot qu'on recherche, il peut être changé
texte_min = texte.lower() #on met le texte en minuscule
liste_mots = texte_min.split() #on le divise en liste de mots. La coupe se fait au niveau des espaces
nombre_mots = len(liste_mots) #on compte le nombre de mots
print(f"Il y a {nombre_mots} mots")
````

**Et pour tester la présence du mot recherché:**

````python

for mot in liste_mots:
    if mot == mot_recherche:
        compteur = compteur + 1
print(f"Il y a {compteur} fois le mot {mot_recherche}")
````

En fait, il y a même plus simple avec la fonction `.count(mot_recherche)`

````python
freq_mot_recherche = texte_min.count("livre")
print(f"Il y a {freq_mot_recherche} fois le mot {mot_recherche}")
````

## Niveau 3 : Les expressions régulières

La difficulté, quand on recherche un terme dans un texte, est qu'il peut exister plusieur graphies : minuscules, majuscules, pluriel, féminin, formes adjectivales, verbales ou nominales. Par exemple, je peux vouloir retrouvrer toutes les fois qu'on parle de participation, que ce soit sous la forme "participatif.ve" ou "participation". 

Une façon commode de résoudre le problème est d'utiliser des **regex** ou **expressions régulières**.

Pour cela, on utilise le module `re` que l'on doit importer :



````python
import re
````

Les fonctions de recherche proposées par le module `re` sont :

- `.findall(pattern, string)` retourne toutes les occurrences du motif (pattern) présent dans le texte exploré (string)
- `.search(pattern, string)` retourne la première occurence du motif (pattern) présent dans le texte exploré (string)
- `.match(pattern, string)` retourne le match si le motif (pattern) se trouve au début du texte (string)

Les regex peuvent par ailleurs être utilisé dans dans un document word ou libre office.

```
Il existe de nombreuses ressources sur le web pour comprendre le fonctionnement des regex et le rôle des différentes "balises". Vous pouvez en particulier consulter le tutoriel sur le site W3school : [https://www.w3schools.com/python/python_regex.asp](https://www.w3schools.com/python/python_regex.asp)
```

````python
print(re.search(r"participati\w+", texte))
print(re.findall(r"participati\w+", texte))

test = "participative, participatives, test participation, participat"
````

Lorsqu'on utilise une expression plusieurs fois, la documentation du module `re` conseille de la "compiler" en utilisant la fonction `re.compile(pattern, flag*)`
Il est possible d'ajouter des "flags" pour ignorer la casse par exemple

````python
reg = re.compile(r"participat\w+", flags= re.IGNORECASE)
reg.findall(test)
````

## Niveau 4: tagger un terme

````python
for n, l in enumerate(lines):
    for w in set(reg.findall(l)):
        l = re.sub(w, f'**{w}**', l.lower())
    lines[n] = l
````

## Niveau 5 : anonymisation/pseudonymisation

Quand on travaille avec des entretiens, il est souvent nécessaire de les pseudonomyser. On peut pour cela utiliser l'outil "find and replace" de son logiciel de traitement de texte. On peut également le faire avec python, notamment lorsqu'on a de nombreux noms de personnes ou de lieux à remplacer.

C'est ce que propose de faire la fonction ci-dessous

````python
# On crée un dictionnaire de pseudonyme
dict_pseudo = {"Speaker 0" : "Enquêté",
               "Speaker 1" : "Enquêteur"}

# on crée une copie de la liste des lignes du texte pour ne pas toucher à  l'original
lines_rec = lines.copy()

for line in lines : #pour chaque ligne de l'entretien
    index_line = lines.index(line) #on récupère la position de la ligne dans la liste "lines" (son index)
    for pseudo in dict_pseudo : #pour chaque pseudo dans le dictionnaire
        new_line = re.sub(pseudo, dict_pseudo[pseudo], line) #on remplace le nom par son pseudo
    lines_rec[index_line] = new_line #on remplace la ligne par la ligne pseudonymisée dans la liste "lines_rec"
    
````

On voit dans ci-dessous que "Speaker 1" a été remplacé par "Enquêteur"

````python
print(lines[0])
print(lines_rec[0])
````
