```python
button = widgets.Button(description="Click Me to Start!")
output = widgets.Output()

i = 0



def on_button_clicked(b):
    global i
    with output:
        button.description = "Next!"
        extract = itw[float(rows[i]['start_in_ms'])*1000:float(rows[i]['end_in_ms'])*1000]
        display(extract)
        new_speaker = widgets.Text(
                        placeholder='Speaker name',
                        description='Speaker:',
                        disabled=False   
                    )
        new_text = widgets.Textarea(
            value=rows[i]['text'].strip(),
            placeholder='Type something',
            description='String:',
            disabled=False,
            layout=widgets.Layout(width='50%', height='80px')
            )
        display(new_speaker)
        display(new_text)
        if os.path.exists("../records/2026-02-02_itw_borrini_cleaned.csv"):
            with open("../records/2026-02-02_itw_borrini_cleaned.csv" , 'a', newline='') as csvfile:
                fieldnames = ['id','start', 'end','cleaned_speaker', 'text','cleaned_text', 'start_in_ms','end_in_ms']
                writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
                writer.writerow({'id': rows[i]['id'],
                                 'start': rows[i]["start"],
                                 'end': rows[i]["end"],
                                 'cleaned_speaker':new_speaker.value,
                                 'text':rows[i]['text'].strip(),
                                 'cleaned_text' : new_text.value,
                                 'start_in_ms':rows[i]['start_in_ms'],
                                 'end_in_ms':rows[i]['end_in_ms']})
        else:
            with open("../records/2026-02-02_itw_borrini_cleaned.csv" , 'w', newline='') as csvfile:
                fieldnames = ['id','start', 'end','cleaned_speaker', 'text','cleaned_text', 'start_in_ms','end_in_ms']
                writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
                writer.writeheader()
                writer.writerow({'id': rows[i]['id'],
                                 'start': rows[i]["start"], 
                                 'end': rows[i]["end"], 
                                 'cleaned_speaker':new_speaker.value, 
                                 'text':rows[i]['text'].strip(), 
                                 'cleaned_text' : new_text.value, 
                                 'start_in_ms':rows[i]['start_in_ms'], 
                                 'end_in_ms':rows[i]['end_in_ms']})
            
        i+=1


display(button, output)

button.on_click(on_button_clicked)
```
