# Introduction à Python


Un usage classique de Python consiste à créer un "script" que l'on éxecute ensuite en "ligne de commande". Ici, on va l'utiliser avec Jupyter pour donner un aspect plus didactique. Avec Jupyter, le résultat produit par une "ligne" de code est visible immédiatement.

Quand on programme en python, on manipule des "objets" et il existe plusieurs types. Les plus courant, ceux qu'on va utiliser sont :

- Les objets "numériques" (nombres) qui peuvent être des *integers* (nombres entiers : 1) ou des *float* (nombres relatifs: 1.5).
- des chaines de caractères appelés *string*. Elles s'écrivent entre "guillemets"
- des listes : ["pommes", "poires", "fraises"]
- des dictionnaires dans lequel une clé correspond à une valeur : {"nom":"Luneau","prénom":"Aymeric","âge":38}


```python
print(type(1))
print(type(1.5))
print(type("aymeric"))
print(type(["pommes", "poires", "fraises"]))
print(type({"nom":"Luneau","prénom":"Aymeric","âge":38}))
```

Dans la cellule ci-dessus, on fait appelle à 2 fonctions :
 - la fonction `type()` qui donne le type de l'objet mis entre parenthèse
 - la fonction `print()` qui affiche le résultat dans la console

En faisant `print(type(objet quelconque))`, on "demande" à Python d'imprimer dans cellule de sortie le type de l'objet. Par exemple, si j'oublie de mettre "print", jupyter affiche uniquement le type de la dernière ligne.

```python
type(1)
type(1.5)
```

## Les variables

Quand on développe du code, on en arrive vite à manipuler des dizaines d'objets. On utilise alors des variables pour faciliter leur manipulation, mais aussi les modifier.
Pour créer une variable, on écrit:
`ǹom_variable = objet representé`


```python
prenom = "aymeric"
nom = "Luneau"
print("Le contenu de la variable prenom est ", prenom)
age = 38
print("Le contenu de la variable âge est ", age)
```

On peut ensuite faire des opérations sur des variables. Par exemple, en additionnant deux variable représentant des chaînes de caractère, on les concatène.

```python
prenom_nom = prenom + nom
print("Le contenu de la variable prenom_nom est ", prenom_nom)
age = 39
print("Le contenu de la variable âge est maintenant", age)
```

## Les conditions et les boucles

L'utilisation de conditions et de boucles est un préalable à toute programmation.

### Les conditions
La définition d'une condition commence par "if" et se termine avec "else" :

```python
x = input("entrez le nom d'un fruit : ")
if x == "pommes":
    print("Ceci est un fruit")
else:
    print("je ne sais pas ce que c'est")

    
```

On remarquera que le concept d'égalité est représenté par le double signe égal : `==`

### Les boucles

Une boucle permet de tester une condition sur une série de variable. C'est le début de l'automatisation
Il y a plusieurs façon de créer une boucle. Ici on voit uniquement les boucles avec `for`.


```python
for nom in list_fruit:
    print(nom)
```


On reprend notre condition ci-dessus, mais au lieu demander à l'utilisateur d'entrer une valeur, on va prendre notre liste de fruits

```python
for x in list_fruit: #Littéralement, pour chaque élément "x" dans la liste list_fruit:
    if x != "pommes": # si x est égale à pomme
        print("Ceci n'est pas une pomme")
    else: #sinon
        print("Ceci est une pomme")
```

Et si on veut créer une liste des légumes dans mon dictionnaire :

```python
list_legume = []
for nom in mon_dico:
    if mon_dico[nom] == "légume":
        list_legume.append(nom)
    else:
        print(f"{nom} n'est pas un légume. Je ne le met pas dans la liste")

print(list_legume)
```

```python
list_legume = [nom for nom in mon_dico if mon_dico[nom] == "légume"] # la même chose que la cellule précédente mais en raccourci

print(list_legume)
```
