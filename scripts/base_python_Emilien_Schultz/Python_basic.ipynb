# Introduction à Python


Un usage classique de Python consiste à créer un "script" que l'on éxecute ensuite en "ligne de commande". Ici, on va l'utiliser avec Jupyter pour donner un aspect plus didactique. Avec Jupyter, le résultat produit par une "ligne" de code est visible immédiatement.

Quand on programme en python, on manipule des "objets" et il existe plusieurs types. Les plus courant, ceux qu'on va utiliser sont :

- Les objets "numériques" (nombres) qui peuvent être des *integers* (nombres entiers : 1) ou des *float* (nombres relatifs: 1.5).
- des chaines de caractères appelés *string*. Elles s'écrivent entre "guillemets"
- des listes : ["pommes", "poires", "fraises"]
- des dictionnaires dans lequel une clé correspond à une valeur : {"nom":"Luneau","prénom":"Aymeric","âge":38}


```python
print(type(1))
print(type(1.5))
print(type("aymeric"))
print(type(["pommes", "poires", "fraises"]))
print(type({"nom":"Luneau","prénom":"Aymeric","âge":38}))
```

Dans la cellule ci-dessus, on fait appelle à 2 fonctions :
 - la fonction `type()` qui donne le type de l'objet mis entre parenthèse
 - la fonction `print()` qui affiche le résultat dans la console

En faisant `print(type(objet quelconque))`, on "demande" à Python d'imprimer dans cellule de sortie le type de l'objet. Par exemple, si j'oublie de mettre "print", jupyter affiche uniquement le type de la dernière ligne.

```python
type(1)
type(1.5)
```

## Niveau 1 : présence d'un mot dans un texte

```python
compteur = 0
mot_recherche = "python"
phrase = input("Écrivez une phrase: ")
phrase_min = phrase.lower()
liste_mots = phrase_min.split(" ")
nombre_mots = len(liste_mots)
print("Il y a %d mots" % nombre_mots)
for mot in liste_mots:
    if mot == mot_recherche:
        compteur = compteur + 1
print("Il y a %d fois le mot Python"%compteur)
```

## Niveau 3 : regex

on veut coder les avis sur python avec une balise <avis> </avis> inspirée du xml.

- On utilise une fonction : petit tout de ce qu'est une fonction
- On veut créer un outil facile à utiliser : on fait notre fonction


```python
import re
```

```python
texte = "Ceci est une phrase de <avis>test</avis> entre nous"
re.findall("<avis>(.*?)</avis>",texte)
```

On crée une fonction

```python
def recup_avis(texte):
    return re.findall("<avis>(.*?)</avis>",texte)
```

```python
recup_avis("ceci est un texte et <avis> je ne suis pas d'accord</avis>")
```

Et on peut l'appliquer sur notre corpus

```python
with open("./entretien Guido von Rossum.txt","r") as f:
    texte = f.read()
recup_avis(texte)
```
